# Summary
Example of how loading pre-computed pickle files can be done.
Special attention is made to store the files in the data directory.

```python
import pickle 
import src.config
import pandas as pd

output_dir = src.config.DATA_DIRECTORY / 'processed' 
output_dir.mkdir(exist_ok = True)
feature_pickle = output_dir / 'feature_table.pickle'
all_tiles_pickle = output_dir / 'full_tileset.pickle'
```

```python
with open(all_tiles_pickle,'rb') as f:
    full_data = pickle.load(f)
```

```python
with open(feature_pickle, 'rb') as f:
    features = pickle.load(f)
```
